```python
def split_text(full_text, chunk_size):
    encoder_model = tiktoken.encoding_for_model('gpt-4o')
    total_encoding = encoder_model.encode(full_text)
    total_token_count = len(total_encoding)

    text_list= []
    for i in range(0, total_token_count, chunk_size):
        chunk = total_encoding[i: i+chunk_size]
        decoding= encoding_model.decode(chunk)
        text_list.append(decoding)
    
    return text_list
```

```python
chunk_list = split_text(full_text, 1500)
chunk_list
```

# 3. 벡터 데이터베이스에 저장

```python
# %pip install chromadb
```

```python
## 임베딩 모델 지정
import os
from chromadb.utils.embedding_functions import OpenAIEmbeddingFunction

OPEN_API_KEY = os.getenv('OPENAI_API_KEY')
openai_embedding = OpenAIEmbeddingFunction(
    api_key=OPEN_API_KEY,
    model_name = 'text-embedding-3-large'
)
```

```python
## 크로마 클라이언트 생성
import chromadb
chroma_client = chromadb.Client()
```

```python
# collection_name = 'law_collection'

# law_collection = chroma_client.get_or_create_collection(
#     collection_name, 
#     embedding_function=openai_embedding,
# )

collection_name = 'law_collection' #저장할 공간 만듦듦
law_collection = chroma_client.get_or_create_collection(collection_name, embedding_function=openai_embedding)
```

```python
## 컬렉션에 설정할 인덱스 생성
id_list = []

for index in range(len(chunk_list)): #분할된 문서 # 청크개수만큼 인덱스 생성
    id_list.append(f'{index}')

## 길이 확인

len(id_list), len(chunk_list)
```

```python
law_collection.add(documents=chunk_list, ids=id_list) #분할된 문서, 인덱스
```

# 4. 질문이 있으면, 벡터 데이터베이스에서 유사도 검색

# 5. 질문 - > 답변

```python
from openai import OpenAI
client = OpenAI()

response = client.responses.create( #메서드
    model="gpt-4o", #모델지정
    instructions=f''' 
    - 당신은 전세사기피해 법률 전문가입니다.
    - [context]를 참고하여사용자의 질문에 답변해주세요.
    [context]
    {retrieved_doc['documents'][0][0]}
    ''', #역할부여
    input=query,#질문
)

print(response.output_text) #openai에서 만든 변수 output_text
```
